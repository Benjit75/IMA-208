# Panorama

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Wed Feb  2 13:49:38 2022

@author: roux
"""


import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
from matplotlib.pyplot import imshow, show, subplot, title, axis, figure


# draw matches
def draw_matches(img1, kpt1, img2, kpt2, matches):
    h1, w1 = img1.shape[:2]
    h2, w2 = img2.shape[:2]

    # Create a blank image with the size of the first image + second image
    new_img = np.zeros((max([h1, h2]), w1 + w2, 3), dtype='uint8')
    new_img[:h1, :w1, :] = np.dstack([img1, img1, img1])
    new_img[:h2, w1:w1 + w2, :] = np.dstack([img2, img2, img2])

    # extract the match keypoints
    for m in matches:
        (x1, y1) = kpt1[m.queryIdx].pt
        (x2, y2) = kpt2[m.trainIdx].pt

        # Draw circles on the keypoints
        cv.circle(new_img, (int(x1), int(y1)), 4, (0, 255, 0), 1)
        cv.circle(new_img, (int(x2) + w1, int(y2)), 4, (0, 255, 0), 1)

        # Connect the same keypoints
        cv.line(new_img, (int(x1), int(y1)), (int(x2) + w1, int(y2)), (255, 0, 255), 1)

    return new_img 

def warpImages(img1, img2, M):
    # get the corner coordinates of the "query" and "train" image
    h1, w1 = img1.shape[:2]
    h2, w2 = img2.shape[:2]
    pts_corners_src = np.float32([[0, 0], [0, h1], [w1, h1], [w1, 0]]).reshape(-1, 1, 2)
    pts_corners_temp = np.float32([[0, 0], [0, h2], [w2, h2], [w2, 0]]).reshape(-1, 1, 2)

    # When we have established a homography we need to warp perspective
    # perform perspective tranform using previously calculated matrix and the corners of  
    #"query" image#
    # Change field of view
    pts_corners_dst = cv.perspectiveTransform(pts_corners_temp, M)

    ListOfPoints = np.concatenate((pts_corners_src, pts_corners_dst), axis=0)

    [x_min, y_min] = np.int32(ListOfPoints.min(axis=0).ravel() - 0.5)
    [x_max, y_max] = np.int32(ListOfPoints.max(axis=0).ravel() + 0.5)

    print(x_min,y_min,x_max,y_max)

    translation_dist = [-x_min, -y_min]

    H_translation = np.array([[1, 0, translation_dist[0]], [0, 1, translation_dist[1]], [0, 0, 
1]])

    new_img = cv.warpPerspective(img2, H_translation.dot(M), (x_max - x_min, y_max - y_min))
    new_img[translation_dist[1]:h1 + translation_dist[1], translation_dist[0]:w1 + translation_dist[0]] = img1

    return new_img



def find_homography(kpt1, kpt2, matches):
    # Find an Homography matrix between two pictures
    # Transforming keypoints to list of points
    src_pts = np.float32([kpt1[m.queryIdx].pt for m in matches]).reshape(-1, 1, 2)
    dst_pts = np.float32([kpt2[m.trainIdx].pt for m in matches]).reshape(-1, 1, 2)

    # Compute a rigid transformation (without depth, only scale + rotation + translation) /affine transformation
    transformation_rigid_matrix, mask = cv.estimateAffinePartial2D(src_pts, dst_pts)
    # transformation_rigid_matrix, mask = cv.estimateAffinePartial2D(src_pts, dst_pts)

    print(transformation_rigid_matrix)
    
    affine_row = [0, 0, 1]
    M= np.vstack((transformation_rigid_matrix, affine_row))
        
    M, mask = cv.findHomography(src_pts, dst_pts,cv.RANSAC,3)
    matchesMask = mask.ravel().tolist()
    print(M)
    
    #return transformation_rigid_matrix,matchesMask

    return M

def test_pano_with_feature(filename1='palaiseau-B1sub.tif', filename2='palaiseau-B2sub.tif', feature_type='hog'):
    outputName=f'panorama-{filename1.split(".")[0]}.tif'

    img1c = cv.imread(filename1,1)
    img2c = cv.imread(filename2,1)
    #
    #img2c = cv.imread('panorama.tif',1)
    #img1c = cv.imread('palaiseau-B3sub.tif', 1)

    img1 = cv.cvtColor(img1c, cv.COLOR_BGR2GRAY)
    img2 = cv.cvtColor(img2c, cv.COLOR_BGR2GRAY)

    img1c = cv.cvtColor(img1c, cv.COLOR_BGR2RGB)
    img2c = cv.cvtColor(img2c, cv.COLOR_BGR2RGB)

    # Initiate ORB detector, that it will be our detector object.
    if feature_type == 'hog':
        model = cv.ORB_create(10000)
    elif feature_type == 'sift':
        model = cv2.xfeatures2d.SIFT_create(10000)
    elif feature_type == 'surf':
        model = cv2.xfeatures2d.SURF_create(10000)

    # find the keypoints and compute the descriptors with ORB for images
    kpts1, des1 = model.detectAndCompute(img1, None)
    kpts2, des2 = model.detectAndCompute(img2, None)


    # Create a BFMatcher object.
    bf = cv.BFMatcher_create(cv.NORM_HAMMING)

    # match descriptor
    matches1to2 = bf.knnMatch(des1, des2, k=2)


    # draw matches
    good1to2 = []
    for m, n in matches1to2:
        if m.distance < 0.6 * n.distance:
            good1to2.append(m)

    drawimg=draw_matches(img1, kpts1, img2, kpts2, good1to2)


    # find  affine transformation and panoramic view between 1 to 2
    matrix1to2 = find_homography(kpts1, kpts2, good1to2)
    img1to2 = warpImages(img2c, img1c, matrix1to2)

    figure()

    imshow(drawimg)
    show()


    figure()

    subplot(1, 2, 1)
    imshow(img1c)
    axis('off')
    title('1')

    subplot(1, 2, 2)
    imshow(img2c)
    axis('off')
    title('2')

    show()

    figure(figsize=(10, 6))

    title1 = f"panoramic between img1 and img2 feature {feature_type}"

    subplot(1, 1, 1)
    imshow(img1to2)
    axis('off')
    title(title1)

    show()

    img1to2 = cv.cvtColor(img1to2, cv.COLOR_BGR2RGB)
    cv.imwrite(outputName,img1to2)
```

**Answers to questions 1 over `panorama.py`:**

- Panorama.py implements homography-based transformations for image alignment and stitching. Homography is a 3x3 transformation matrix that maps the pixels of one image to the corresponding pixels in another image, accounting for perspective distortion and camera motion. This transformation is appropriate for stitching images that have significant overlap and are captured from the same viewpoint.

- Panorama.py uses the HOG (Histogram of Oriented Gradients) feature descriptor for identifying keypoints and matching them across images. There are several other feature descriptors that can be used instead of HOG, such as SIFT, SURF, just by changing the `feature_type` parameter when calling `test_pano_with_feature()`.

- The distance threshold in Panorama.py controls the maximum distance between matched keypoints that are considered as valid correspondences. If the threshold is too high, then more keypoints will be matched, but some of them may be incorrect matches due to noise or other factors. If the threshold is too low, then fewer keypoints will be matched, but the matches will be more reliable.

# Corners

```python
import cv2 
import numpy as np 
import matplotlib.pyplot as plt

def test_corners(filename="f7.tif", get_corners=False):
    img = cv2.imread(filename) 
    imrvb=img.copy()

    gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)     

    kernel = np.ones((9,9),np.uint8)
    gray = cv2.GaussianBlur(gray,(9,9),0)

    edges = cv2.Canny(gray,20,180,apertureSize =3) 

    lines = cv2.HoughLines(edges,1,2*np.pi/180,100) 

    nbrlines = lines.shape[0]
    
    if not get_corners:
        plt.figure()
        plt.imshow(gray)
        plt.show()
        plt.figure('Canny edges')
        plt.imshow(edges)
        plt.show()
        print('nombre de lignes : ', nbrlines)

    corners = []
    for line1 in range(nbrlines):  
        r1,theta1 = lines[line1][0]
        for line2 in range(line1):
            r2,theta2 = lines[line2][0]

            a1 = np.cos(theta1) 
            b1 = np.sin(theta1)
            c1 = -r1
            a2 = np.cos(theta2) 
            b2 = np.sin(theta2)
            c2 = -r2

            X =   b1 * c2 - c1 * b2
            Y =   c1 * a2 - a1 * c2
            Z =   a1 * b2 - b1 * a2

            if Z != 0:
                X=X/Z
                Y=Y/Z   
                if 0 < X < img.shape[1] and 0 < Y < img.shape[0]:
                    corners.append((int(X),int(Y)))
                    x1=int(X+5)
                    y1=int(Y+5)
                    x2=int(X-5)
                    y2=int(Y-5)

                    a=255
                    b=0
                    c=0

                    cv2.line(imrvb,(x1,y1), (x2,y2), (a,b,c),5) 
                    cv2.line(edges,(x1,y1), (x2,y2), (a,b,c),5) 

                    x1=int(X-5)
                    y1=int(Y+5)
                    x2=int(X+5)
                    y2=int(Y-5)

                    cv2.line(imrvb,(x1,y1), (x2,y2), (a,b,c),5) 
                    cv2.line(edges,(x1,y1), (x2,y2), (a,b,c),5) 

    if not get_corners:
        plt.figure()
        plt.title('edges')
        plt.imshow(edges)
        plt.show()

        plt.figure()
        plt.title('image')
        plt.imshow(imrvb)
        plt.show()

    return corners if get_corners else None
```

**Answers of questions 2 over `corner.py`:**

- The purpose of the Gaussian blur is to smooth out the image and reduce the effects of noise, thus improving the accuracy of the edge detection performed by the Canny algorithm.
- Other filtering techniques that could have been more appropriate for this specific task include median filtering, which is often used for reducing the effect of salt-and-pepper noise, or bilateral filtering, which can preserve edges while smoothing out the rest of the image.
- The main elements of the script are:
    - Reading and loading the input image using OpenCV
    - Converting the image to grayscale
    - Applying a Gaussian blur to the grayscale image
    - Detecting edges using the Canny algorithm
    - Applying the Hough transform to detect lines in the edge image
    - Finding intersections between the detected lines to determine the corners of the image
    - Drawing lines and corners on the original image and displaying the results
- The Canny algorithm is used to detect edges in the image. It takes three parameters: the input image, a lower threshold value, and an upper threshold value. Pixels with gradient values above the upper threshold are considered edges, while pixels below the lower threshold are not edges. Pixels with gradient values between the two thresholds are only considered edges if they are connected to pixels above the upper threshold. The Hough transform is used to detect lines in the edge image. It takes four parameters: the input image, the resolution of the rho parameter, the resolution of the theta parameter, and a threshold value. The rho parameter is the distance from the origin to the line, and the theta parameter is the angle between the line and the x-axis. The threshold value determines the minimum number of votes required for a line to be considered valid. Finally, the intersections between the detected lines are found to determine the corners of the image.
